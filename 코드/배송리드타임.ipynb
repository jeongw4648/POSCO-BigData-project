```python
# 데이터 구성하기
df_sales = pd.read_csv('C:/Users/Administrator/Desktop/A2_유아용품/Sales_data.csv')
df_sales
```

```python
# 데이터 구성하기
df_member = pd.read_csv('C:/Users/Administrator/Desktop/A2_유아용품/Member_data.csv')
df_member
```

```python
# 결측치 제거
df_member.isnull().sum()
```

```python
df_member.dropna(inplace=True)
```

```python
df_member
```

```python
# 변수 타입 변경
df_sales['구매일'] = pd.to_datetime(df_sales['구매일'])
df_sales['배송시작일'] = pd.to_datetime(df_sales['배송시작일'])
df_sales['배송완료일'] = pd.to_datetime(df_sales['배송완료일'])
print(df_sales.dtypes)
```

```python
# 배송 리드 타임 계산
df_sales['배송리드타임'] = df_sales['배송완료일']-df_sales['구매일']
df_sales
```

```python
# 배송리드타임 변수 타입 변경
leadtime = df_sales['배송리드타임'].dt.days
df_sales['배송리드타임'] = leadtime
df_sales
```

```python
s6 = df_sales.groupby(by = ['고객번호'])['배송리드타임'].sum()
```

```python
df_member = pd.merge(df_member, s6, how = 'left', on = '고객번호')
df_member
```

```python
# 배송리드타임 변수 중에서 5일 이상 걸린 주문 건수
lsr = df_sales['배송리드타임']
a = 0
for i in lsr:
    if i > 4:
        a += 1
print(a, '개')
```

```python
# 배송리드타임이 5일 이상 걸린 확률(배송지연율)
num = 13943/199999*100
print(round(num, 2), '%')
```
