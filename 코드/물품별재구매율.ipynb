# 물품 별 재구매 횟수 

```python
products_list = list(df_products['상품명'])
```

```python
products_dic = {}
for product in products_list:
    if product in products_dic:
        products_dic[product+'2'] = 0
    else: products_dic[product] = 0
```

```python
repurchase_dic = {}
for idx, sale in df_sales.iterrows():
    if sale['고객번호'] not in repurchase_dic:
        repurchase_dic[sale['고객번호']] = {}
    if sale['상품명'] not in repurchase_dic[sale['고객번호']]:
        repurchase_dic[sale['고객번호']][sale['상품명']] = 1
    else: repurchase_dic[sale['고객번호']][sale['상품명']] += 1   
```

```python
repurchase_copy_dic = copy.deepcopy(repurchase_dic)
for custom_key in repurchase_copy_dic:
    for product_key in repurchase_copy_dic[custom_key]:
        if repurchase_dic[custom_key][product_key] == 1:
            del(repurchase_dic[custom_key][product_key])
repurchase_dic   
```

```python
for custom_key in repurchase_dic:
    for product_key in repurchase_dic[custom_key]:
        products_dic[product_key] += 1
#products_dic
```

```python
for product_key in products_dic:
    if products_dic[product_key] != 0:
        products_dic[product_key] -=1
#products_dic
```

```python
df_products
```

```python
df_products['재구매횟수'] = list(products_dic.values())
df_products['재구매비율'] = (df_products['재구매횟수']/df_products['물품판매량']).round(3)
```

```python
df_products.to_csv('물품별재구매율.csv', index=False, encoding="utf-8-sig") 
```

```python
from datetime import datetime

def spliting_date(df, data_row):
    df[data_row] = pd.to_datetime(df[data_row])
    df[data_row+'_연도'] = df[data_row].dt.year
    df[data_row+'_월'] = df[data_row].dt.month
    df[data_row+'_일'] = df[data_row].dt.day
    df[data_row+'_요일'] = df[data_row].dt.day_name()
```
