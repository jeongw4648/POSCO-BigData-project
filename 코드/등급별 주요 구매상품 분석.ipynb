```python
df_1_count = df_raw_LEFT_JOIN[df_raw_LEFT_JOIN.grade == '1등급'].groupby('물품중분류')['주문번호'].agg(['count']).sort_values('count', ascending=False)
df_1_count['count'] = (df_1_count['count'] / df_1_count['count'].sum()).round(3)
df_1_count.head(5)
```

```python
df_not_1_count = df_raw_LEFT_JOIN[df_raw_LEFT_JOIN.grade != '1등급'].groupby('물품중분류')['주문번호'].agg(['count']).sort_values('count', ascending=False)
df_not_1_count['count'] = (df_not_1_count['count'] / df_not_1_count['count'].sum()).round(3)
df_not_1_count.head(5)
```
