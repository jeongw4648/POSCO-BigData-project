```python
from datetime import datetime
df_sales['구매일'] = pd.to_datetime(df_sales['구매일'])
df_sales['배송시작일'] = pd.to_datetime(df_sales['배송시작일'])
df_sales['배송완료일'] = pd.to_datetime(df_sales['배송완료일'])
```

```python
df_sales['구매일_연도'] = df_sales['구매일'].dt.year
df_sales['구매일_월'] = df_sales['구매일'].dt.month
df_sales['구매일_일'] = df_sales['구매일'].dt.day
df_sales['구매일_요일'] = df_sales['구매일'].dt.day_name()
```

```python
# 데이터 구성하기
df_product = pd.read_csv('C:/Users/Administrator/Desktop/엑셀파일/Product_분류(대분류,중분류,브랜드).csv',encoding='euc-kr')
df_product
```

```python
z2 = df_sales.groupby(by = ['상품명'])['배송리드타임'].sum()
z3 = df_sales.groupby(by = ['상품명'])['배송리드타임'].count()

total_avg_readtime = (z2 / z3).round(0)
df_product = pd.merge(df_product, total_avg_readtime, how = 'left', on = '상품명')
df_product.rename(columns={'배송리드타임':'평균 배송리드타임'},inplace=True)
df_product
```

```python
z4 = df_sales.groupby(by = ['상품명'])['할인금액'].sum()
z5 = df_sales.groupby(by = ['상품명'])['할인금액'].count()

total_avg_sale = (z4 / z5).round(0)

df_product = pd.merge(df_product, total_avg_sale, how = 'left', on = '상품명')
df_product.rename(columns={'할인금액':'평균 할인금액'},inplace=True)
df_product
```
